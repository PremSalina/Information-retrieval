```python
import wikipedia
import json

wikipedia.set_lang("en")

def scrape_data_education(target_documents=510):
    documents = []

    # Define a list of education-related keywords
    education_keywords = ["Education", "Literacy rates", "Online education trends", "Student loan data",
                          "Educational institutions", "Curriculum development", "Teaching methods", "Educational technology",
                          "Education policies", "Education statistics", "E-learning", "Distance education",
                          "Educational research", "Teacher training", "Educational assessments", "Education reform",
                          "Higher education", "K-12 education", "Special education", "Adult education"]

    for keyword in education_keywords:
        search_results = wikipedia.search(keyword, results=target_documents)

        for title in search_results:
            try:
                content = wikipedia.page(title, auto_suggest=False)
                if len(documents) >= target_documents:
                    break

                if any(kw in content.summary.lower() for kw in education_keywords) and len(content.summary) >= 200:
                    document_info = {
                        'topic': "Education",
                        'title': content.title,
                        'revision_id': content.revision_id,
                        'summary': content.summary,
                        'url': content.url,
                    }
                    documents.append(document_info)

                    if len(documents) >= target_documents:
                        break

            except Exception as e:
                print(f"An error occurred: {e}")

        if len(documents) >= target_documents:
            break

    return documents

# Scrape education-related data
education_doc_list = scrape_data_education(target_documents=600)

# Save the scraped data to a JSON file
json_filename = "education_scraped_data.json"

with open(json_filename, 'w') as json_file:
    json.dump(education_doc_list, json_file, indent=5)

print(f"Scraped {len(education_doc_list)} documents related to education and saved them to '{json_filename}'")
```
