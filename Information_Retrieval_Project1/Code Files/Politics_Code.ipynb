```python
import wikipedia
import json

wikipedia.set_lang("en")

def scrape_data_politics(target_documents=510):
    documents = []

    # Define a list of politics-related keywords
    politics_keywords = ["Politics", "Elections", "Public policy analysis", "International relations",
                        "Political systems", "Government policies", "Political ideologies", "Political leaders",
                        "Political campaigns", "Political parties", "Legislation", "Political activism",
                        "Political science", "Foreign policy", "Domestic policy", "Political history",
                        "Political theory", "Public administration", "Political events", "Political movements"]

    for keyword in politics_keywords:
        search_results = wikipedia.search(keyword, results=target_documents)

        for title in search_results:
            try:
                content = wikipedia.page(title, auto_suggest=False)
                if len(documents) >= target_documents:
                    break

                if any(kw in content.summary.lower() for kw in politics_keywords) and len(content.summary) >= 200:
                    document_info = {
                        'topic': "Politics",
                        'title': content.title,
                        'revision_id': content.revision_id,
                        'summary': content.summary,
                        'url': content.url,
                    }
                    documents.append(document_info)

                    if len(documents) >= target_documents:
                        break

            except Exception as e:
                print(f"An error occurred: {e}")

        if len(documents) >= target_documents:
            break

    return documents

# Scrape politics-related data
politics_doc_list = scrape_data_politics(target_documents=600)

# Save the scraped data to a JSON file
json_filename = "politics_scraped_data.json"

with open(json_filename, 'w') as json_file:
    json.dump(politics_doc_list, json_file, indent=5)

print(f"Scraped {len(politics_doc_list)} documents related to politics and saved them to '{json_filename}'")
```
