```python
import wikipedia
import json

wikipedia.set_lang("en")

def scrape_data_technology(target_documents=510):
    documents = []

    # Define a list of technology-related keywords
    technology_keywords = ["Technology", "Emerging technologies", "AI advancements", "Artificial intelligence",
                           "Machine learning", "Robotics", "Automation", "Internet of Things (IoT)",
                           "Cybersecurity", "Big data", "Blockchain", "Augmented reality (AR)",
                           "Virtual reality (VR)", "Quantum computing", "Nanotechnology", "Biotechnology",
                           "Information technology", "Computer science", "Innovation", "Tech startups"]

    for keyword in technology_keywords:
        search_results = wikipedia.search(keyword, results=target_documents)

        for title in search_results:
            try:
                content = wikipedia.page(title, auto_suggest=False)
                if len(documents) >= target_documents:
                    break

                if any(kw in content.summary.lower() for kw in technology_keywords) and len(content.summary) >= 200:
                    document_info = {
                        'topic': "Technology",
                        'title': content.title,
                        'revision_id': content.revision_id,
                        'summary': content.summary,
                        'url': content.url,
                    }
                    documents.append(document_info)

                    if len(documents) >= target_documents:
                        break

            except Exception as e:
                print(f"An error occurred: {e}")

        if len(documents) >= target_documents:
            break

    return documents

# Scrape technology-related data
technology_doc_list = scrape_data_technology(target_documents=600)

# Save the scraped data to a JSON file
json_filename = "technology_scraped_data.json"

with open(json_filename, 'w') as json_file:
    json.dump(technology_doc_list, json_file, indent=5)

print(f"Scraped {len(technology_doc_list)} documents related to technology and saved them to '{json_filename}'")
```
