```python
import wikipedia
import json

wikipedia.set_lang("en")

def scrape_data_health(target_doc=510):
    doc= []


    health_key = ["health", "common diseases", "global health statistics",
                      "mental health trends", "public health", "healthcare",
                      "medical research", "nutrition", "wellness", "health education",
                      "vaccination", "epidemics", "pandemics", "healthcare systems",
                      "health policies", "health initiatives", "health campaigns"]

    for key in health_key:
        result = wikipedia.search(key, results=target_doc)

        for title in result:
            try:
                content = wikipedia.page(title, auto_suggest=False)
                if len(doc) >= target_doc:
                    break

                if any(k in content.summary.lower() for k in health_key) and len(content.summary) >= 200:
                    doc_info = {
                        'topic': "Health",
                        'title': content.title,
                        'revision_id': content.revision_id,
                        'summary': content.summary,
                        'url': content.url,
                    }
                    doc.append(doc_info)

                    if len(doc) >= target_doc:
                        break

            except Exception as exp:
                print(f"An error occurred: {exp}")

        if len(doc) >= target_doc:
            break

    return doc

doc_list = scrape_data_health(target_doc=510)

json_fname = "health_scrap_data.json"

with open(json_fname, 'w') as file_json:
    json.dump(doc_list, file_json, indent=5)

print(f"Scraped {len(doc_list)} documents related to health and saved them to '{json_fname}'")
```
