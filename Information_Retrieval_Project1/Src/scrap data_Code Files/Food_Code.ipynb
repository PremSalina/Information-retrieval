```python
import wikipedia
import json

wikipedia.set_lang("en")

def scrape_data_food(target_documents=510):
    documents = []

    # Define a list of food-related keywords
    food_keywords = ["Food", "Crop yield statistics", "Global hunger", "Food security",
                     "Agriculture", "Food production", "Nutrition", "Food distribution",
                     "Food supply chain", "Food access", "Food sustainability",
                     "Food waste", "Foodborne diseases", "Hunger relief",
                     "Food policy", "Food assistance programs"]

    for keyword in food_keywords:
        search_results = wikipedia.search(keyword, results=target_documents)

        for title in search_results:
            try:
                content = wikipedia.page(title, auto_suggest=False)
                if len(documents) >= target_documents:
                    break

                if any(kw in content.summary.lower() for kw in food_keywords) and len(content.summary) >= 200:
                    document_info = {
                        'topic': "Food",
                        'title': content.title,
                        'revision_id': content.revision_id,
                        'summary': content.summary,
                        'url': content.url,
                    }
                    documents.append(document_info)

                    if len(documents) >= target_documents:
                        break

            except Exception as e:
                print(f"An error occurred: {e}")

        if len(documents) >= target_documents:
            break

    return documents

# Scrape food-related data
food_doc_list = scrape_data_food(target_documents=600)

# Save the scraped data to a JSON file
json_filename = "food_scraped_data.json"

with open(json_filename, 'w') as json_file:
    json.dump(food_doc_list, json_file, indent=5)

print(f"Scraped {len(food_doc_list)} documents related to food and saved them to '{json_filename}'")
```
